# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_names = school_data_complete["school_name"].unique()
len(school_names)
# school_count = school_data_complete["school_name"].value_counts()
# school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["Student ID"].unique()
len(student_count)
```

```python
# Calculate the total budget
total_budget = school_data_complete["budget"].unique()
sum(total_budget)
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_count
```

```python
passing_math_percentage = '{:.2%}'.format(passing_math_count / len(student_count))
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_count
```

```python
passing_reading_percentage = '{:.2%}'.format(passing_reading_count/len(student_count))
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = '{:.2%}'.format(passing_math_reading_count/len(student_count))
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

district_summary = {'Total Schools':[len(school_names)],\
                    'Total Students':[len(student_count)],\
                    'Total Budget':[sum(total_budget)],\
                    'Average Math Score': [average_math_score],\
                    'Average Reading Score':[average_reading_score],\
                    '% Passing Math':[passing_math_percentage],\
                    '% Passing Reading':[passing_reading_percentage],\
                    '% Overall Passing':[overall_passing_rate]}

district_summary = pd.DataFrame(district_summary)

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
school_type = school_data_complete.groupby(["school_name"])["type"].unique()
school_type = pd.DataFrame(school_type)
school_type.columns=["Type"]
school_type
```

```python
# Calculate the total student count
per_school_counts = school_data_complete.groupby(["school_name"])["student_name"].count()
per_school_counts = pd.DataFrame(per_school_counts)
per_school_counts.columns=["Student Count"]
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"])["budget"].mean()
per_school_budget = pd.DataFrame(per_school_budget)
per_school_budget.columns=["School Budget"]
per_school_budget
```

```python
per_school_capita = per_school_budget["School Budget"]/per_school_counts["Student Count"]
per_school_capita = pd.DataFrame(per_school_capita)
per_school_capita.columns=["School Capita"]
per_school_capita
```

```python
# Calculate the average test scores
school_math_avg = school_data_complete.groupby(["school_name"])["math_score"].mean()
school_math_avg = pd.DataFrame(school_math_avg)
school_math_avg.columns=["Average Math Score"]
school_math_avg
```

```python
school_reading_avg = school_data_complete.groupby(["school_name"])["reading_score"].mean()
school_reading_avg = pd.DataFrame(school_reading_avg)
school_reading_avg.columns=["Average Reading Score"]
school_reading_avg
```

```python
# Calculate the number of students per school with math scores of 70 or higher
stu_passing_math = school_data_complete[school_data_complete["math_score"]>=70]
stu_passing_math = stu_passing_math.groupby(["school_name"])["math_score"].count()
stu_passing_math = pd.DataFrame(stu_passing_math)
stu_passing_math.columns=["# Students Passing Math"]
stu_passing_math
```

```python
# Calculate the number of students with reading scores of 70 or higher
stu_passing_reading = school_data_complete[school_data_complete["reading_score"]>=70]
stu_passing_reading = stu_passing_reading.groupby(["school_name"])["reading_score"].count()
stu_passing_reading = pd.DataFrame(stu_passing_reading)
stu_passing_reading.columns=["# Students Passing Reading"]
stu_passing_reading
```

```python
# Use the provided code to calculate the students that passed both math and reading with scores of 70 or higher
stu_passing_math_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
stu_passing_math_reading = stu_passing_math_reading.groupby(["school_name"])["student_name"].count()
stu_passing_math_reading = pd.DataFrame(stu_passing_math_reading)
stu_passing_math_reading.columns=["# Students Passing Overall"]
stu_passing_math_reading
```

```python
# Use the provided code to calculate the passing rates
rate_passing_math = stu_passing_math["# Students Passing Math"]/per_school_counts["Student Count"]
rate_passing_math = pd.DataFrame(rate_passing_math)
rate_passing_math.columns=["% Students Passing Math"]
rate_passing_math
```

```python
rate_passing_reading = stu_passing_reading["# Students Passing Reading"]/per_school_counts["Student Count"]
rate_passing_reading = pd.DataFrame(rate_passing_reading)
rate_passing_reading.columns=["% Students Passing Reading"]
rate_passing_reading
```

```python
overall_passing_rate = stu_passing_math_reading["# Students Passing Overall"] / per_school_counts["Student Count"]
overall_passing_rate = pd.DataFrame(overall_passing_rate)
overall_passing_rate.columns=["% Students Overall Passing"]
overall_passing_rate
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

school_summary = pd.concat([school_type, per_school_counts, per_school_budget, per_school_capita, school_math_avg, school_reading_avg, rate_passing_math, rate_passing_reading, overall_passing_rate], axis=1)

school_summary = pd.DataFrame(school_summary)

# Formatting
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:,.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:,.2f}".format)
school_summary["% Students Passing Math"] = school_summary["% Students Passing Math"].map('{:.2%}'.format)
school_summary["% Students Passing Reading"] = school_summary["% Students Passing Reading"].map('{:.2%}'.format)
school_summary["% Students Overall Passing"] = school_summary["% Students Overall Passing"].map('{:.2%}'.format)
school_summary["School Budget"] = school_summary["School Budget"].map("${:,.2f}".format)
school_summary["School Capita"] = school_summary["School Capita"].map("${:,.2f}".format)

# # Display the DataFrame
# per_school_summary
school_summary
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary.copy()
school_spending_df["School Capita"] = school_spending_df["School Capita"].str.replace('\W','',regex=True)
school_spending_df["% Students Passing Math"] = school_spending_df["% Students Passing Math"].str.replace('\W','',regex=True)
school_spending_df["% Students Passing Reading"] = school_spending_df["% Students Passing Reading"].str.replace('\W','',regex=True)
school_spending_df["% Students Overall Passing"] = school_spending_df["% Students Overall Passing"].str.replace('\W','',regex=True)
school_spending_df["School Capita"] = pd.to_numeric(school_spending_df["School Capita"])
school_spending_df["% Students Passing Math"] = pd.to_numeric(school_spending_df["% Students Passing Math"])
school_spending_df["% Students Passing Reading"] = pd.to_numeric(school_spending_df["% Students Passing Reading"])
school_spending_df["% Students Overall Passing"] = pd.to_numeric(school_spending_df["% Students Overall Passing"])
school_spending_df["School Capita"] = school_spending_df["School Capita"].floordiv(100)
school_spending_df["% Students Passing Math"] = school_spending_df["% Students Passing Math"].floordiv(100)
school_spending_df["% Students Passing Reading"] = school_spending_df["% Students Passing Reading"].floordiv(100)
school_spending_df["% Students Overall Passing"] = school_spending_df["% Students Overall Passing"].floordiv(100)
# The above code is to remove the previous formatting so I could do the binning later.
school_spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["School Capita"], bins=spending_bins, labels=labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["School Capita"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["School Capita"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["School Capita"])["% Students Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["School Capita"])["% Students Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["School Capita"])["% Students Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.concat([spending_math_scores, spending_reading_scores, spending_passing_math, spending_passing_reading, overall_passing_spending], axis=1)

# Display results
spending_summary
```
